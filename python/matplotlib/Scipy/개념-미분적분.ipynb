```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
%matplotlib widget
a = sp.symbols('alpha')
sol = sp.solve(2*a + 1)
print(sol[0])
print(sol)

x = np.linspace(-2,2,1000)
fig = plt.figure()
ax = fig.add_subplot()
ax.axhline(0,color='lightgray')
ax.axvline(0,color='lightgray')
ax.axvline(sp.solve(2*a + 1),color='lightgray', linestyle='--')
ax.axhline((a**2 +a).subs(a,sp.solve(2*a + 1)[0]),color='red', linestyle='--')
ax.plot(x,2*x + 1)
ax.plot(x,x**2+x)
```

# Newton Kenetic Equatiton
$$
\ddot{X_i} = \frac{X_{i+1} - 2 x_{i} + X_{i-1}}{ \tau^2}  = \frac{F(X_i)}{m}
$$

$$
X_{i-1} = \frac{\tau^2}{m}\,F(X_i) + 2X_i - X_{i-1}
