# [Line integral](https://www.youtube.com/watch?v=yyqOvA8dlvk&list=PL4ElPQw0dPdlhkDrkVLsktCZWkILzKhvM&index=3)
> Filed F by line r
    $$ \int_{c} \vec{F} \cdot d\vec{r} \\
        \int_{c} \vec{F}(\vec{r}(t)) \cdot \frac{dr}{dt}dt
    $$
    

# Problem

> vector field 
> - $\vec u \,= \,(xy, z^2, x) $

> curve
> - $x = 1 + t , y = 0, z = t^2, \\ 
    0 \leq x \leq 3 $

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
%matplotlib widget

t = np.linspace(0,3,100)

rx = 1 + t
rz = np.zeros_like(rx)
ry = t**2
x = np.linspace(0,5,10)
x,y,z = np.meshgrid(x,x,x)
ux = x*y
uy = z**2
uz = x

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_title(r'$\vec{u} = < xy,z^2,x>, curve = <1+t,0,t2>, 0 \leq x \leq 3$')
#ax.quiver(x,y,z,ux,uy,uz,length=0.1,normalize=True)
ax.plot(rx,ry,rz,'r')
plt.show()

```

# solve

$$ \int_{c} \vec{u}(\vec{r(t)}) \cdot \frac{dr}{dt}dt \\
    \vec{u} = < xy, z^2, x >, \\
    \vec{u} = < (1+t)* 0, (t^2)^2, 1+t > \\
    curve = \vec {r} = < 1+t, 0, t^2 > \to \vec{r}dt = < 1,0,2t> \\
    < 0,t^4,1+t> \cdot <1,0,2t>dt \to < 0,0,2t + t^2>dt \\
    \int_{t_0}^{t_3} <0,0,2t+t^2>dt \\
        = \int_{0}^{3} 2t+t^2 dt \\
        = \left . t^2 + \frac{1}{3}t^3 \right|_{0}^{3}
$$
