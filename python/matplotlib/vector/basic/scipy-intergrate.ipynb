# Fourier Transforms
    - The discrete Fourier transform is defined as
    
$$ y[k] = \sum_{n=0}^{N-1} e^{-2{\pi}in(k/N)}\,x[n]$$

    - k / N represents a specific frequency (dimensionless)
    - can be converted to a frequency(Hz) if you know the spacing in x

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.cos(x) + 1j*np.sin(x)
eix = np.exp(1j*x)
eix_ = np.exp(-1j*x)
#plt.plot(x,y)
#plt.plot(x,eix+1,label=r'$e^i$')
plt.plot(x,np.sin(x),label=r'$\sin(x)$')
plt.plot(x,np.sin(3*x),label=r'$\sin(3x)$')
plt.plot(x,np.sin(-3*x),label=r'$\sin(-3x)$')
#plt.plot(x,np.exp(-1j*x),label=r'$e^-i$')
plt.legend()
```
