```python
import numpy as np
import sympy as sm
import matplotlib.pyplot as plt
%matplotlib widget
x,y,z=sm.symbols('x y z',real=True)
#def f(x,y,z):
#    return 5*(x-3) + 1*(y-4) + 3*(z-2)
f = 5*(x-3) + (y-4) + 3*(z-2)
f=sm.solve(f,z)
#z=sm.solve(f(x,y,z),z)
f=sm.lambdify((x,y),f[0],'numpy')
x = np.linspace(-1,10,1000)
X,Y = np.meshgrid(x,x)
fig = plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter3D([3],[4],[2])
ax.plot_surface(X,X,f(X,Y))
ax.quiver([3],[4],[2],[5],[1],[3],color='r')
f(3,4)
```
