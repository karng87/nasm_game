# heaviside 
> ## numpy.heaviside(t,0)  
> ## numpy.heaviside(t-1,0)

```python
import numpy as np
import matplotlib as npl
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
def x(t):
    # heaviside step function( if x1 = 0 then x2 )
    return np.heaviside(t,0) - np.heaviside(t-1,0)
```

```python
t = np.linspace(-5,5,1000)

fig = plt.figure(figsize=(6,6))

ax = fig.add_subplot(511)
ax.plot(t,np.heaviside(t,0),label='h(t,0)')
ax.plot(t,np.heaviside(t-1,0),label='h(t-1,0)')
plt.legend()

ax = fig.add_subplot(512)
ax.plot(t,np.heaviside(t,0)-np.heaviside(t-1,0),label='h(t,0)-h(t-1,0)')
plt.legend()

ax = fig.add_subplot(513)
ax.plot(t,t*np.heaviside(t,0),label='t*h(t,0)')
plt.legend()

ax = fig.add_subplot(514)
ax.plot(t,t*np.heaviside(t-1,0),label='t*h(t-1,0)')
plt.legend()

ax = fig.add_subplot(515)
ax.plot(t,t*np.heaviside(t,0)-t*np.heaviside(t-1,0),label=r'$t\cdot h(t,0) - t\cdot h(t-1,0)$')
plt.legend()
```
