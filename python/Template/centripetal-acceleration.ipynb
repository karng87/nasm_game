> ### $
\Delta{v}^2 = v^2 + v^2 - 2v^2 cos\theta
\\ 
\Delta{v} = \sqrt{2v^2(1-cost\theta)} = \sqrt{2}\,v \,\sqrt{1-cos\theta}\\
\frac{\Delta{v}}{\Delta{t}} = \frac{\sqrt{2}\,v \,\sqrt{1-cos\theta}}{t}
\frac{\Delta{v}}{\Delta{t}} = \frac{\sqrt{2}\,v\omega \,\sqrt{1-cos\theta}}{t}
$

> ### $
\displaystyle \lim_{t\to 0} \frac{\Delta{v}}{\Delta{t}} = \frac{\sqrt{2}\,v \,\sqrt{1-cos\theta}}{\frac{\theta}{\omega}}
= \displaystyle \lim_{\theta \to 0}\frac{\sqrt{2}\,v\omega \,\sqrt{1-cos\theta}}{\theta}
= \displaystyle \lim_{\theta \to 0} \frac{\sqrt{2}\,v\omega \,\sqrt{1-cos\theta}}{\theta} 
\Big(\frac{\sqrt{1+cos\theta}}{\sqrt{1+cos\theta}} \Big)
\\
\therefore \displaystyle \lim_{\theta \to 0} \frac{ \sqrt{2}\,v\omega \,\sin\theta}
{\theta \sqrt{1+cos\theta}}
= \displaystyle \lim_{\theta \to 0} \frac{\sin\theta}{\theta}
\displaystyle \lim_{\theta \to 0} \frac{\sqrt{2} v \omega}{\sqrt{1+cos\theta}}
= v \omega \\
\because \omega = \frac{v}{r}, v = r\omega \quad a = \frac{v^2}{r} = r \omega^2
\therefore F_{(구심력)}\; = m\frac{v^2}{r} = mr\omega^2 
$
