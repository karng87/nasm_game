```python
import plotly 
import numpy as np

x,y = np.meshgrid(np.arange(0,2,0.2),
                  np.arange(0,2,0.2)
                 )
u = np.cos(x)
v = np.sin(y)

fig = plotly.figure_factory.create_quiver(x,y,u,v)
fig.show()
```

```python
x,y = np.meshgrid(np.arange(-2,2,0.2),
                  np.arange(-2,2,0.25))
z = x * np.exp(-x**2 - y**2)
v,u = np.gradient(z,0.2,0.2)

fig = plotly.figure_factory.create_quiver(x,y,u,v,
                                          scale=0.25,
                                          arrow_scale=0.4,
                                          name='quiver',
                                          line_width=1
                                         )
fig.add_trace(plotly.graph_objects.Scatter(x=[-0.7,0.7],
                                           y=[0,0],
                                           mode='markers',
                                           marker_size=12,
                                           name='points')
                                          )
fig.show()
```
