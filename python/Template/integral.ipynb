# Picard Integration Method

```python
import numpy as np
import sympy as sym
import sympy.utilities
import scipy as sci
import scipy.optimize

import matplotlib.pyplot as plt
import matplotlib.patches as pat 

x,y,alpha, beta = sym.symbols('x y \alpha \beta')
F = lambda x: (x+2)**2 - 1

f = sym.diff(F(x), x)
fx = sym.lambdify(x,f)

print(F(-3))
print(F(1))
sn = sym.integrate(f,(x,-4,-3))

a0 = -4
b0 = -3
px = np.linspace(a0,b0)
py = sym.lambdify(x,f)(px)
pa = [(a0,0), *zip(px,py), (b0,0)]
po = pat.Polygon(pa)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
xi = np.linspace(-5,5,1000)
ax.plot(xi,F(xi))

#a = sci.optimize.fmin_bfgs(F,0)
#print (a)
#ax.anotate('(%s, %s)' %xy, xy)
ax.plot(xi,fx(xi))
ax.add_patch(po)
ax.text(0.2*a0+b0, -2, sn)
ax.plot(-4,F(-4),'o')
ax.text(-4,F(-4),"(-4, " + str(F(-4)) +")")

a2 = np.linspace(-2,1,100)
a3 = F(a2)
ax.fill_between(a2,a3)
ax.axhline(0,color='gray')
ax.axvline(0,color='gray')
```

```python
import numpy as np
import sympy as sym
import scipy as sci

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import warnings
warnings.filterwarnings("ignore") 

x,y = sym.symbols('x y')
an = sym.solve(sym.ln(x),x)
ans = sym.solveset(sym.ln(x),x)
it = iter(ans)
print(next(it))

xi = np.linspace(0.01,10,1000)
yd = [sym.ln(x) for x in xi]

fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlabel('$x$')
ax.set_ylabel("$f(x)$")
ax.set_xlim(xmin=-5, xmax=5)
ax.set_ylim(ymin=-5, ymax=5)
ax.axvline(0)
ax.axhline(0)
ax.axvline(an)
ax.grid()
ax.plot(xi,yd, color='orange')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon


def func(x):
    return (x - 3) * (x - 5) * (x - 7) + 85


a, b = 2, 9  # integral limits
x = np.linspace(0, 10)
y = func(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.set_ylim(bottom=0)

# Make the shaded region
ix = np.linspace(a, b)
iy = func(ix)
verts = [(a, 0), *zip(ix, iy), (b, 0)]
print("verts: ", verts)
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)

ax.text(0.5 * (a + b), 30, r"$\int_a^b f(x)\mathrm{d}x$",
        horizontalalignment='center', fontsize=20)

fig.text(0.9, 0.05, '$x$')
fig.text(0.1, 0.9, '$y$')

ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([])

plt.show()
```
