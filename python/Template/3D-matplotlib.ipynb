```python
# line eqation
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget
t = np.linspace(-3,3,10)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_aspect('auto',adjustable='box')
ax.plot([0,2],[0,1],[0,1]) # (2,1,1) - (0,0,0)
# x=1+2t, y=1+t, z=0
ax.plot(1+2*t, 1+t, 0*t) # t = (x-1)/2 = (y-1)/1 , z = oo
```

```python
# plane equation
# direction vector = partial differencial

import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_zlim(-3,3)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
xi = np.linspace(-1,3,100)
x,y = np.meshgrid(xi,xi)

ax.quiver(0,0,0,2,1,0,color='r',arrow_length_ratio=0.1)
ax.contour(x, y, 2*(x-1)+(y-1), np.linspace(-3,3,100),alpha=0.5)
ax.plot_surface(x, y, 2*x+y,alpha=0.5)
```

```python
import matplotlib.pyplot as plt 
import matplotlib.patches as pat 
import numpy as np 
%matplotlib widget

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set(xlabel='x',ylabel='y')
x = np.linspace(-3,3)
x,y = np.meshgrid(x,x)
z = 4 - x**2 - y**2
ax.plot_surface(x,y,z,alpha=0.3)
ax.scatter3D([1],[1],[2], c='black')

### f(x,y,z)|{x=1}
t = np.linspace(-3,3)
xi = np.ones_like(t)
yi = t
lx  = 4 - xi**2 - yi**2
ax.plot3D(xi,yi,lx, lw=2,c='green')
### f(x,y,z)|{y=1}
xi = t
yi = np.ones_like(t)
ly = lx
ax.plot3D(xi,yi,ly, lw=2,c='blue')
# Dx(4 - x^2 - y^2) = -2x
# f_x(1) = -2
# f(1,1) = 4
xy = np.linspace(-3,3)
xx = np.ones_like(xy)
xz = -2*xy + 4 
ax.plot3D(xx,xy,xz,lw=2,c='red')

# Dy(4 - x^2 - y^2) = -2y
# f_y(1) = -2
# f(1,1) = 4
yx = t
yy = np.ones_like(t)
yz = -2*yx + 4
ax.plot3D(yx,yy,yz,lw=2,c='cyan')
```
