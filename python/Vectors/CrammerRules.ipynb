# Cramer's rules

```python
import numpy as np
import sympy as sm
import sympy.vector
N = sm.vector.CoordSys3D('N')

# v(3,1), w(1,2) \iff T([3,1],[1,2])
u = 3*N.i + N.j + 0*N.k
v = N.i + 2*N.j + 0*N.k
w = 0*N.i + 0*N.j + N.k
T = u.to_matrix(N).col_insert(1,v.to_matrix(N)).col_insert(2,w.to_matrix(N))
T = u.to_matrix(N).row_join(v.to_matrix(N)).row_join(w.to_matrix(N))
B = T.inv()
# x(x_0, x_1)
x0,x1,x2 = sm.symbols('x_0 x_1 x_2')
x = sm.Matrix([x0,x1,x2])
# sm.vector.matrix_to_vector(x,N)

# Tx = h(3,2)
h = 3*N.i + 2*N.j + 0*N.k

import matplotlib.pyplot as plt
%matplotlib widget
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot()
ax.grid()
ax.spines[['left','bottom']].set_position('zero')
ax.set_aspect('equal')
ax.set_xlim(xmin=-2,xmax=10)
ax.set_ylim(ymin=-2,ymax=10)
ax.set_xticks(np.arange(-2,10,1))
ax.set_yticks(np.arange(-2,10,1))
ax.quiver(0,0,3,1,width=0.05,scale=1,units='xy')
ax.quiver(0,0,-1/5,3/5,width=0.02,scale=1,units='xy')
ax.annotate(r"$A_{i,1}=\vec{e_1}(3,1)$",(3.1,0.8))
ax.quiver(0,0,1,2,width=0.05,scale=1,units='xy')
ax.quiver(0,0,2/5,-1/5,width=0.02,scale=1,units='xy')
ax.annotate(r'$A_{i,2}=\vec{e_2}(1,2)$',(0.1,2.5))
ax.annotate(r"${A^{-1}}_{i,1}=\vec{e_1}(\frac{2}{5},-\frac{1}{5})$",(-0.5,-0.7))
ax.annotate(r"${A^{-1}}_{i,2}=\vec{e_2}(-\frac{1}{5},\frac{3}{5})$",(-2,0.7))
ax.quiver(0,0,4,3,width=0.02,scale=1,units='xy')
ax.annotate(r'$\vec{b}(4,3)$',(4.1,3))
ax.plot([1,4],[2,2],color='r')
ax.plot([1,4],[2,3],color='k')
ax.plot([4,4],[2,3],color='r')
ax.plot([3,3],[0,1],color='r')
ax.plot([0,3],[0,0],color='r')

ax.plot([3,4],[1,3],color='k')
ax.plot([3,4],[3,3],color='b')
ax.plot([0,0],[0,2],color='b')
ax.plot([0,1],[2,2],color='b')
ax.plot([3,3],[2,3],color='b')
ax.plot([3,3],[1,2],color='b')
ax.plot([0,1],[1,1],color='g')
ax.plot([1,1],[1,0],color='g')
ax.plot([3,12],[1,4],'k--',linewidth=0.5)
ax.plot([1,5],[2,10],'k--',linewidth=0.5)
ax.plot([4,8],[3,11],'k--',linewidth=0.5)
ax.plot([6,10],[2,10],'k--',linewidth=0.5)
ax.plot([9,10],[3,5],'k--',linewidth=0.5)

ax.plot([4,10],[3,5],'k--',linewidth=0.5)
ax.plot([2,11],[4,7],'k--',linewidth=0.5)
ax.plot([3,12],[6,9],'k--',linewidth=0.5)
ax.plot([4,10],[8,10],'k--',linewidth=0.5)
ax.plot(6,7,'ro')
ax.quiver(0,0,6,7,width=0.05,scale=1,units='xy')
ax.plot([1,7],[2,9],'r--',linewidth=0.5)
ax.plot([6,7],[7,9],'r--',linewidth=0.5)

ax.plot([3,9],[1,8],'g--',linewidth=0.5)
ax.plot([6,9],[7,8],'g--',linewidth=0.5)

# solve
T.inv()*sm.Matrix([3,2,0])
(T.inv()*x).T*(h.to_matrix(N))
B
T
```
