```python
t = np.linspace(0,100,301)
def f(t):
    return np.array([t+t/10*np.sin(t),t+(t/10.)*np.cos(t)])
t1 = np.linspace(0,30,200)
t2 = np.linspace(0,50,200)
t3 = np.linspace(0,90,200)

import matplotlib.pyplot as plt
%matplotlib widget
import numpy as np
fig,ax = plt.subplots(1,3,figsize=(10,3),sharey=True)
ax[0].plot(f(t1)[0],f(t1)[1],label=r'$t=30$ s',color='b')
ax[1].plot(f(t2)[0],f(t2)[1],label=r'$t=50$ s',color='b')
ax[2].plot(f(t3)[0],f(t3)[1],label=r'$t=90$ s',color='b')
ax[0].quiver(0,0,f(t1)[0,-1],f(t1)[1,-1],scale=1,scale_units='xy')
ax[1].quiver(0,0,f(t2)[0,-1],f(t2)[1,-1],scale=1,scale_units='xy')
ax[2].quiver(0,0,f(t3)[0,-1],f(t3)[1,-1],scale=1,scale_units='xy')

for i in ax:
    i.set_xlim(0,100)
    i.set_ylim(0,100)
    i.legend(loc="upper left")

ax[1].set_xlabel('y')
ax[0].set_ylabel('y')
```

```python
def r(t):
    return np.array([np.cos(t),np.sin(t)])
def dr(t):
    return np.array([-np.sin(t),np.cos(t)])

t = np.linspace(0,2*np.pi,100)
plt.figure(figsize=(5,5))
plt.plot(1.4*np.cos(t),1.4*np.sin(t),color='b')
plt.plot(r(t)[0],r(t)[1],color='k',lw=2)
plt.plot(dr(t)[0],r(t)[1],color='r',lw=2)
plt.quiver(r(0.5)[0],r(0.5)[1],
           dr(0.5)[0],
           dr(0.5)[1],
           scale_units='xy',color='r'
          )
plt.quiver(r(3.8)[0],r(3.8)[1],
           dr(3.8)[0],
           dr(3.8)[1],
           scale_units='xy',color='r'
          )
plt.scatter(0,0)
plt.quiver(0,0,r(0.5)[0],r(0.5)[1],scale=1,scale_units='xy')
plt.quiver(0,0,r(3.8)[0],r(3.8)[1],scale=1,scale_units='xy')
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
r(t)
```

```python
import sympy as sm
import sympy.vector
R = sm.vector.CoordSys3D('')
Omega = R.x**2 + R.x*R.y + R.y**2
display(Omega)
display(sm.vector.gradient(Omega))
plt.figure()
plt.gca().set_aspect('equal')
x=np.linspace(0,10,10)
y=np.linspace(0,10,10)
X,Y = np.meshgrid(x,y)
Omega = X**2 + X*Y + Y**2
plt.contour(X,Y,Omega,20,cmap='jet',zorder=1,linewidths=2)
plt.quiver(X,Y,2*X+Y,X+2*Y,zorder=2,label=r'$\nabla \Omega$')
plt.colorbar(label=r'$\Omega=x^2+xy+y^2$',fraction=0.046,pad=0.04)
plt.legend(loc='lower left')
```
