```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
np.random.seed(32)

mu = 3
n = 50

x = np.random.normal(mu, 1, size=n)
y = np.random.normal(mu, 1, size=n)
z = np.random.normal(mu, 1, size=n)
```

```python
plt.rcParams['figure.figsize'] = (8,6)
```

# 3D Scatter Plot

```python
ax = plt.axes(projection='3d')
ax.scatter3D(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(45,20)
# 1st arg elevation
# 2nd arg rotation
```

```python
type(ax)
```

```python
omega = 2
z_line = np.linspace(0,10,100)
x_line = np.cos(omega*z_line)
y_line = np.sin(omega*z_line)
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(x_line, y_line, z_line, lw=4)
```

# Surface

```python
f = lambda x, y: np.sin(x) * np.cos(y)
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X, Y = np.meshgrid(x,y)
F = f(X,Y)

ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,F)
plt.show()
```

# Interactive View

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import cm # color map

f = lambda x, y: np.sin(x) * np.cos(y)
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X, Y = np.meshgrid(x,y)
F = f(X,Y)

def func(A, E):
    ax = plt.axes(projection = '3d')
    ax.plot_surface(X,Y,F, cmap=cm.coolwarm)
    ax.view_init(elev=E, azim=A)
    plt.rcParams['figure.figsize'] = (8,6)
    plt.show()
    
#f_interactive(45,45)

from ipywidgets import interactive
iplot = interactive(func,E=(-90,90,5), A=(-90,90,5))
iplot
```
