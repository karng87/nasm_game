# Total Derivative( $df  = \nabla \cdot f$ )
cf. Taylor Series => sympy.sin(x).series()
> ### $$
\begin{align}
f(x,y,z)|_{a,b,c} & = 
f(a,b,c) \\ & + 
 \frac{\partial f}{\partial x}(x-a) + 
 \frac{\partial f}{\partial y}(y-a) + 
 \frac{\partial f}{\partial z}(z-a) \\ & + 
\frac{(x-a)^2}{2!}\frac{\partial ^2f}{\partial x^2} + 
\frac{(y-a)^2}{2!}\frac{\partial ^2f}{\partial y^2} +
\frac{(z-a)^2}{2!}\frac{\partial ^2f}{\partial z^2} + ...
\end{align}
$$

> ## $df = \lim \limits_{x\to a\\ y\to b \\ z \to c}(f(x,y,z) - f(a,b,c))$
>> ### $ 
df = 
\begin{cases}
\frac{\partial f}{\partial x}(x-a) + 
\frac{\partial f}{\partial y}(y-a) + 
\frac{\partial f}{\partial z}(z-a)  
\\ 
\quad \therefore
\frac{\partial f}{\partial x}(dx) + 
\frac{\partial f}{\partial y}(dy) + 
\frac{\partial f}{\partial z}(dz)  
\\
(\frac{\partial f}{\partial x} ,
\frac{\partial f}{\partial y} ,
\frac{\partial f}{\partial z}) \cdot (dx,dy,dz)
\end{cases}
\\
\quad \therefore \nabla f = 
(\frac{\partial f}{\partial x} ,
\frac{\partial f}{\partial y} ,
\frac{\partial f}{\partial z}) 
\\
\quad \because d\vec{r} = (dx, dy, dz)
= (\frac{\partial f}{\partial x} ,
\frac{\partial f}{\partial y} ,
\frac{\partial f}{\partial z}) 
\cdot 
(dx, dy, dz) 
\\
= (\frac{\partial}{\partial x}f ,
\frac{\partial}{\partial y}f ,
\frac{\partial}{\partial z}f) 
\cdot 
(dx, dy, dz) 
\\
= (\frac{\partial}{\partial x} ,
\frac{\partial}{\partial y} ,
\frac{\partial}{\partial z})f 
\cdot 
(dx, dy, dz) 
\\
= \nabla f \cdot (dx, dy, dz) 
\\
= \nabla f \cdot d \, \vec{r} 
$

# Directional derivative
> ### $
\begin{align}
\because
df & = (
\frac{\partial}{\partial x},
\frac{\partial}{\partial y},
\frac{\partial}{\partial z}) 
\,f \cdot d\vec{r} \\
& = (
\frac{\partial}{\partial x},
\frac{\partial}{\partial y},
\frac{\partial}{\partial z}) 
\,f \cdot dl\,\hat r \\
\frac{df}{dl}  & = (
\frac{\partial}{\partial x},
\frac{\partial}{\partial y},
\frac{\partial}{\partial z}) 
\,f \cdot \hat r \\
& = 
\nabla f \cdot \hat r \\
\end{align}
\\
\qquad \mbox{Directional derivative } \overset {def}{\iff} 
\begin{cases}
\nabla f \cdot \vec{\hat r} \\
  |\nabla f| |\hat r| \cos(\theta) \\
  |\nabla f| \cos(\theta) 
  \end{cases}
$

# Prove the third rule
> ## $ \nabla \cdot (f\,\vec{v}) = f(\nabla \cdot \vec{v}) + \vec{v}\cdot (\nabla f)$

````python
nabla = sm.vector.Del()
v1, v2, v3, f = sm.symbols('v_1 v_2 v_3 f',cls=sm.Function)
vF = v1(C.x,C.y,C.z)*C.i + v2(C.x,C.y,C.z)*C.j + v3(C.x,C.y,C.z)*C.k
sF = f(C.x,C.y,C.z)
````

````python
lhs = nabla.dot(sF*vF)
lhs
````

````python
rhs = sF*(nabla.dot(vF)) + vF.dot(nabla(sF))
rhs
````

````python
lhs.doit().expand().simplify() == rhs.doit().expand().simplify()
````
