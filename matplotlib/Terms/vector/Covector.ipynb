# Covector
> ### $ [V^{*},\; \mathbb{R},\; e^m,\; 4]$
>> ### $ 
[V, \mathbb{R}, e_m, 4] \rightarrow [V^{*},\; \mathbb{R},\; e^m,\; 4] \\
\qquad \quad \nwarrow \qquad  \swarrow \\
\qquad [V^{**},\; \mathbb{R},\; e,\; 4]
$
> ### vector
>> ### $[V, \mathbb{R}, e_m, 4] \iff A^{m}e_m$
> ### covector
>> ### $[V{*}, \mathbb{R}, e^m, 4] \iff A_{m}e^{m}$
> ### $ <e^i,e_j> = \delta^{i}_{j}$
> ### $ <B_{\mu}e^{\mu},A^{\nu}e_{\nu}> $
>> ### $\because \text{ linear map} \\
= B_{\mu}A^{\nu}<e^{\mu},e_{\nu}>\;\; 
= \;B_{\mu}A^{\nu}\delta^{\mu}_{\nu}\; 
=\; B_{\mu}A^{\mu}\\
= \sum_{i=1}^{\mu}B_iA^i = B \cdot A$

>>> ## $\therefore$ row vector is a function 
>>>> ### $ f: \vec{V} \mapsto \mathbb{R} $
>> row vector는 선형 연산자 이다.  
>>> operator란 입력값으로 받은 집합에 원소로 결과를 출력해준다.
>>> ## $ nf(\vec{v} + \vec{w}) = f(n\vec{v}) + f(n\vec{w}) \\
\begin{cases}
f(\vec{v} + \vec{w}) = f(\vec{v}) + f(\vec{w})\\
nf(\vec{v}) = f(n\vec{v}) 
\end{cases}$
