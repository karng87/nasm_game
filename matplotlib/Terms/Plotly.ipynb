```python
import plotly.io as pio
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
from plotly.validators.scatter.marker import SymbolValidator
import numpy as np
import pandas as pd
from urllib.request import urlopen
```

```python
iris = px.data.iris()
print(iris)
fig = px.scatter(iris, x='petal_width', y='petal_length',
                color='species',size = 'sepal_length',hover_data=['sepal_width'])
fig.show()
```

```python
tips = px.data.tips()
print(tips)
fig = px.scatter(tips,x='total_bill',y = 'tip',color='sex',facet_col='day')
fig.show()
```

```python
t = np.linspace(0,10,100)
y1 = np.random.randn(100).cumsum()
y2 = np.random.randn(100).cumsum()

fig = go.Figure()
fig.add_trace(go.Scatter(x=t,y=y1,name='red',mode='markers',marker_color='darkred'))
fig.add_trace(go.Scatter(x=t,y=y2,name='blue',mode='markers',marker_color='darkblue'))
fig.update_traces(mode='markers',marker_line_width=1,marker_size=10)
```

```python
gapminder = px.data.gapminder()
gm_2017 = gapminder.query('year==2007')
fig = px.scatter(gm_2017,x='gdpPercap', y='lifeExp',
                 size='pop',color='continent',hover_name='country',
                 log_x=True,size_max=60)
fig.show()
```

```python
x = np.linspace(0,2*np.pi,100)
f = lambda x: np.sin(x)
fig  = go.Figure(data=go.Scatter(x=x,  y=f(x)))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x,y=np.sin(x),
                         #mode='markers',
                         name='sin(x)'))
fig.add_trace(go.Scatter(x=x,y=np.cos(x),
                         #mode='lines',
                         name='cos(x)'))
fig.add_trace(go.Scatter(x=x,y=np.sin(x)*np.cos(x), 
                         #mode='markers+lines',
                         name='sin(x)cos(x)'))
```

```python
import plotly as pl

fig = pl.subplots.make_subplots(rows=1,cols=3,
                               shared_yaxes=True,
                               column_titles=['sin(x)','cos(x)','sin(x)cos(x)'],
                               x_title='x',y_title='y')
fig.add_trace(pl.graph_objects.Scatter(x=x,y=np.sin(x),name='sin(x)'),
             row=1,col=2,
             )
fig.add_trace(pl.graph_objects.Scatter(x=x,y=np.cos(x),name='cos(x)'),
              row=1,col=1)
fig.add_trace(pl.graph_objects.Scatter(x=x,y=np.cos(x)*np.sin(x), name='cos(x)sin(x)'),
              row=1,col=3)
fig.update_layout(yaxis=dict(range=[-2,2]))
fig.show()
```

```python
x,y = np.meshgrid(np.arange(0,2,.2),np.arange(0,2,.2))
u = np.cos(x)
v = np.sin(x)
fig = pl.figure_factory.create_quiver(x,y,u,v)
fig.show()
```

```python
x,y = np.meshgrid(np.arange(-2,2,.2),np.arange(-2,2,.25))
z = x * np.exp(-x**2 - y**2)
#z = x * np.exp(-x)
v,u = np.gradient(z,.2,.2)
fig = pl.figure_factory.create_quiver(x,y,u,v,
                                      scale = .25,
                                      arrow_scale = .4,
                                      name = 'quiver',
                                      line_width=1)
fig.add_trace(pl.graph_objects.Scatter(x=[-0.7,0.7], y =[0,0],
                                       mode='markers',
                                       marker_size=12,
                                       name = 'point'))
fig.show()
```

```python
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)

Y,X = np.meshgrid(x,y)
u = -1 - X**2 + Y
v = 1 + X - Y**2
fig = pl.figure_factory.create_streamline(x,y,u,v,arrow_scale=0.1)
fig.show()
```
