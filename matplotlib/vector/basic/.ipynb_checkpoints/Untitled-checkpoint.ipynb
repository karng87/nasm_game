# vector field
$$ \vec F(x,y) = (\sin(x), \cos(x)) $$
$$ \vec F(x,y) = \sin(x)\hat{x} + \cos(x)\hat{y} $$

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = np.linspace(0,2*np.pi,100)
y = np.linspace(0,2*np.pi,100)
z = np.sin(x)
ax.plot(x,y,z,'ro')

x,y = np.meshgrid(x,y)
z = np.sin(x)
ax.plot_surface(x,y,z)
plt.show()


```
