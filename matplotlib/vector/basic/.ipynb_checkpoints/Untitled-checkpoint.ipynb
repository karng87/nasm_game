```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

# shortcuts

> dd : delete cell

> z : to undo

```python
x = np.linspace(0,20,100)
y = np.exp(-x)
y1 = np.exp(-x/10)
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y)
ax.plot(x,y1)

plt.show()
```
