```python
import matplotlib
matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

## $$ \vec{f}(x,y) = \sin(x)\,\hat{x} + \cos(y)\,\hat{y}$$

```python
x_range = y_range = np.linspace(0, 2*np.pi, 10)
x, y = np.meshgrid(x_range, y_range)

#fx, fy = np.sin(x), np.cos(y)
fx, fy = x, y
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()
Q = ax.quiver(x, y, fx, fy, pivot='middle')
ax.quiverkey(Q, .8, 1.02, 1, "= 1m/s", labelpos="E")
ax.axis("scaled")
ax.set_title("Flow")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.streamplot(x,y,fx,fy, density=2)
ax.axis("scaled")
plt.show()
```

```python
x_range = y_range = np.linspace(0, 2*np.pi, 100)
x, y = np.meshgrid(x_range, y_range)
fx, fy = x, y
#fx, fy = np.sin(x), np.cos(y)
f_mag = np.sqrt(fx**2 + fy**2)

fig = plt.figure()
ax = fig.add_subplot()
ax.imshow(f_mag, origin="lower", extent=(0,2*np.pi, 0, 2*np.pi), cmap="gray")
ax.quiver(x[5::10,::10],y[5::10,::10],fx[5::10,::10],fy[5::10,::10], pivot="middle", color="red")

plt.show()
```

## gradient $$ \vec{\nabla}f(x,y) = \frac{\partial{f}}{\partial{x}}\,\hat{x} + \frac{\partial{f}}{\partial{y}}\,\hat{y}$$

```python
def f(x,y):
    return x + y
    #return np.cos(x)**2 * np.cos(y)**2 
```

```python
x_range = y_range = np.linspace(-np.pi/2, np.pi/2, 101)
x, y = np.meshgrid(x_range, y_range)

#dfdy, dfdx = np.gradient(f(x,y))
dfdx, dfdy = np.gradient(f(x,y))
#dfdy, dfdx = np.gradient(f(x,y), np.pi/100)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.contour(f(x,y), origin="lower", extent=(-np.pi/2, np.pi/2, -np.pi/2, np.pi/2))
ax.quiver(x[5::10,5::10],y[5::10,5::10],dfdx[5::10,5::10],dfdy[5::10,5::10], pivot="middle")
plt.show()
```
