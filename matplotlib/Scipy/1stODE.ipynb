```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

from scipy.integrate import odeint
# ivp = initial value problem
from scipy.integrate import solve_ivp
#? solve_ivp
```

```python
k = 1.0
def fprime(x,y):
    return 3 * x**2 + 12*x - 4
tspan = (-8,4)
F0 = (-120,)
sol = solve_ivp(fprime, tspan, F0)
sol
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(sol.t, sol.y[0], 'b-o')
```

```python
x = np.linspace(*tspan)
sol = solve_ivp(fprime, tspan, F0, t_eval=x)
plt.axhline(0)
plt.axvline(0)
plt.plot(sol.t,sol.y[0],'b-')
plt.show()
```

```python
def event(x,y):
    return y
sol = solve_ivp(fprime, tspan, F0, events=event)
sol
```
